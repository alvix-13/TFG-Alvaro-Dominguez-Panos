```python
import pickle
import numpy as np
import pandas as pd
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
path = "data_test_fwsa.csv"

df = pd.read_csv(path)
```

```python
df_f = df.drop(["TimeStamp", "label"], axis=1)
y_df = df["label"]
```

# RandomForest

```python
modelo_final = pickle.load(open('../MODELOS/FWSAs/modelo_RF.pkl','rb'))
```

```python
predicciones = modelo_final.predict(df_f)
```

```python
y_test_numeric = np.array(y_df)
predicciones_numeric = np.array(predicciones)
```

```python
print(classification_report(y_df, predicciones, zero_division=1))
```

```python
forest_accuracy = accuracy_score(y_df, predicciones)

print(f"El valor de accuracy: {forest_accuracy}")
```

```python
cm = confusion_matrix(y_df, predicciones)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.unique(y_df))
disp.plot()
plt.show()
```
